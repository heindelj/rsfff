```python
import pyscf
from gpu4pyscf.properties import eda
```

```python
basis = "def2-svpd"

water = pyscf.M(
    atom="""
        O   0.000000  -0.000000   0.117400
        H  -0.757000  -0.000000  -0.469600
        H   0.757000   0.000000  -0.469600
    """,
    basis=basis,
    charge=0,
    spin=0,
    verbose=4,
)

chloride = pyscf.M(
    atom="""
        Cl  -1.940555   0.888176  -2.558508
    """,
    basis=basis,
    charge=-1,
    spin=0,
    verbose=4,
)

eda_result, dft_result = eda.eval_ALMO_EDA_2_energies(
    [water, chloride],
    xc="wB97M-V",
)

print("\nEDA components:")
for name, value in eda_result.items():
    print(f"{name:20s}: {value}")

print("\nFragment and supersystem energies:")
print(dft_result)
```
